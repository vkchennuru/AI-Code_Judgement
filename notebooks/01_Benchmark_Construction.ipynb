````python
from pathlib import Path
import pandas as pd
````

````python
PROJECT_ROOT = Path("/content/AI-Code_Judgement")

DATASET_PATH = PROJECT_ROOT / "dataset" / "cs1_problems.csv"

OUTPUT_PATH = PROJECT_ROOT / "dataset" / "cs1_problems_clean.csv"

print(DATASET_PATH)
````

## Step 1: Load the Benchmark Dataset

The benchmark dataset contains introductory programming problems that will be used throughout the experimental pipeline.

This section loads the dataset and confirms that it has been read successfully.

````python
# ============================================================
# Load Benchmark Dataset
# ============================================================

df = pd.read_csv(DATASET_PATH)

print("✓ Dataset loaded successfully.")
print(f"Number of rows    : {df.shape[0]}")
print(f"Number of columns : {df.shape[1]}")
````

````python
# Display first five records

df.head()
````

## Step 2: Dataset Information

This section displays the dataset structure, column names, and data types.

````python
df.info()
````

````python
print("Column Names")

for i, col in enumerate(df.columns, start=1):
    print(f"{i}. {col}")
````

## Step 3: Missing Value Analysis

Checking whether any records contain missing values.

````python
missing_values = df.isnull().sum()

print("Missing Values")

missing_values
````

````python
if missing_values.sum() == 0:
    print("✓ No missing values found.")
else:
    print("⚠ Missing values detected.")
````

## Step 4: Duplicate Record Analysis

Checking for duplicate programming problems.

````python
duplicate_rows = df.duplicated().sum()

print(f"Duplicate Rows : {duplicate_rows}")
````

````python
if duplicate_rows == 0:
    print("✓ No duplicate records found.")
else:
    print("⚠ Duplicate records detected.")
````

## Step 5: Dataset Summary Statistics

````python
df.describe(include="all")
````

This section validates the quality of the benchmark dataset by checking:

````python
# ============================================================
# Benchmark Quality Validation
# ============================================================

print("=" * 60)
print("Benchmark Validation")
print("=" * 60)

print(f"Total Programming Problems : {len(df)}")
print(f"Total Features             : {len(df.columns)}")
````

````python
print("\nDataset Columns\n")

for col in df.columns:
    print("✓", col)
````

## Step 8: Save Clean Benchmark Dataset

The validated benchmark dataset is saved for use in the remaining notebooks.

````python
# ============================================================
# Save Clean Dataset
# ============================================================

df.to_csv(OUTPUT_PATH, index=False)

print("✓ Clean dataset saved successfully.")
print("Output File:")
print(OUTPUT_PATH)
````
