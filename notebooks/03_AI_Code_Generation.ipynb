````python
# ============================================================
# Import Libraries
# ============================================================

import os
import pandas as pd
from pathlib import Path
````

````python
# ============================================================
# Define Project Paths
# ============================================================

PROJECT_ROOT = Path("/content/AI-Code_Judgement")

PROMPTS_DIR = PROJECT_ROOT / "prompts"

GENERATED_CODE_DIR = PROJECT_ROOT / "generated_code"

RESULTS_DIR = PROJECT_ROOT / "results"

GENERATED_CODE_DIR.mkdir(parents=True, exist_ok=True)

RESULTS_DIR.mkdir(parents=True, exist_ok=True)

print("Project Root :", PROJECT_ROOT)
print("Prompt Folder:", PROMPTS_DIR)
print("Generated Code Folder:", GENERATED_CODE_DIR)
````

## Load Prompt Manifest

The prompt manifest created in Notebook 2 maps benchmark problems to their corresponding prompt files.

````python
manifest_path = PROMPTS_DIR / "prompt_manifest.csv"

manifest_df = pd.read_csv(manifest_path)

display(manifest_df.head())

print()

print(f"Total Prompt Files : {len(manifest_df)}")
````

## Identify Prompt Files

The prompt manifest is used to locate all prompt files that will be submitted to the Large Language Model (LLM) during the code generation stage.

````python
# ============================================================
# Locate Prompt Files
# ============================================================

prompt_files = []

for _, row in manifest_df.iterrows():

    prompt_file = PROMPTS_DIR / row["prompt_file"]

    prompt_files.append(prompt_file)

print("=" * 60)
print("Prompt Files Available")
print("=" * 60)

for file in prompt_files:
    print(file.name)

print()
print(f"Total Prompt Files : {len(prompt_files)}")
````

## Manual AI Code Generation Procedure

Each prompt file was independently submitted to the selected Large Language Model (LLM).

For every benchmark problem:

- The same prompt was submitted three independent times.
- Each response was obtained in a fresh interaction to minimize conversational context effects.
- The returned Python programs were saved as:

- `solution_A.py`
- `solution_B.py`
- `solution_C.py`

The generated solutions were stored in the corresponding problem directory under `generated_code/`.

````python
# ============================================================
# Create Problem Directories
# ============================================================

for pid in manifest_df["problem_id"]:

    problem_dir = GENERATED_CODE_DIR / f"problem_{int(pid):03d}"

    problem_dir.mkdir(parents=True, exist_ok=True)

print("✓ Problem directories created successfully.")
````

````python
print("Generated Code Directories\n")

for folder in sorted(GENERATED_CODE_DIR.iterdir()):

    if folder.is_dir():
        print(folder.name)
````

## Organization of Generated Solutions

Each benchmark problem has a dedicated directory with three independently generated Python solutions.

Example:

```
generated_code/
    problem_001/
        solution_A.py
        solution_B.py
        solution_C.py
```

This directory structure is used by Notebook 4 for automated functional evaluation.

## Verify Generated AI Solutions

This section verifies that the three generated Python solutions exist for every benchmark problem before functional evaluation.

````python
# ============================================================
# Verify Generated Solutions
# ============================================================

verification_records = []

for _, row in manifest_df.iterrows():

    pid = int(row["problem_id"])

    problem_dir = GENERATED_CODE_DIR / f"problem_{pid:03d}"

    solution_A = problem_dir / "solution_A.py"
    solution_B = problem_dir / "solution_B.py"
    solution_C = problem_dir / "solution_C.py"

    verification_records.append({
        "problem_id": pid,
        "solution_A": solution_A.exists(),
        "solution_B": solution_B.exists(),
        "solution_C": solution_C.exists()
    })

verification_df = pd.DataFrame(verification_records)

display(verification_df)
````

````python
print("=" * 60)
print("Solution Verification Summary")
print("=" * 60)

print("Total Problems :", len(verification_df))

print()

print("Solutions Available")

print("Solution A :", verification_df["solution_A"].sum())
print("Solution B :", verification_df["solution_B"].sum())
print("Solution C :", verification_df["solution_C"].sum())
````

````python
assert verification_df["solution_A"].all()
assert verification_df["solution_B"].all()
assert verification_df["solution_C"].all()

print("✓ All generated solutions verified successfully.")
````
